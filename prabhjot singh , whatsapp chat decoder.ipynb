```python
# Feature 1 - Chat Parser

messages = []

participants = set()

system_count = 0
media_count = 0
deleted_count = 0

with open("hostel_bois.txt", "r", encoding="utf-8") as file:

    for line in file:

        line = line.strip()

        # Skip empty lines
        if not line:
            continue

        # Check if line contains timestamp separator
        if " - " not in line:
            continue

        # Split timestamp and remaining text
        timestamp, rest = line.split(" - ", 1)

        # System message (no sender)
        if ": " not in rest:
            system_count += 1
            continue

        # Split sender and message
        sender, message = rest.split(": ", 1)

        # Skip media messages
        if message == "<Media omitted>":
            media_count += 1
            continue

        # Skip deleted messages
        if message == "This message was deleted":
            deleted_count += 1
            continue

        # Store parsed message
        messages.append({
            "timestamp": timestamp,
            "sender": sender,
            "message": message
        })

        participants.add(sender)

# Print summary
print("Successfully parsed", len(messages), "messages")
print("Participants :", len(participants))
print("System messages :", system_count)
print("Media omitted :", media_count)
print("Deleted messages :", deleted_count)

print("\nParticipants:")
for person in participants:
    print(person)

print("\nFirst 5 messages:")
for msg in messages[:5]:
    print(msg)
```

```python
from datetime import datetime

# Total messages
total_messages = len(messages)

# Count messages per person
person_count = {}

for msg in messages:
    sender = msg["sender"]

    if sender in person_count:
        person_count[sender] += 1
    else:
        person_count[sender] = 1

# List of participants
participants = sorted(person_count.keys())

# Number of participants
participant_count = len(participants)

# Parse dates
date_format = "%d/%m/%y, %H:%M"

first_date = datetime.strptime(messages[0]["timestamp"], date_format)
last_date = datetime.strptime(messages[-1]["timestamp"], date_format)

days = (last_date - first_date).days + 1

# Sort by message count
sorted_people = sorted(person_count.items(),
                       key=lambda x: x[1],
                       reverse=True)

print("=" * 60)
print("GROUP OVERVIEW".center(60))
print("=" * 60)

print(f"Total Messages : {total_messages}")
print(f"Participants   : {participant_count}")
print(f"Period         : {first_date.strftime('%d %B %Y')} to {last_date.strftime('%d %B %Y')} ({days} days)")

print("\nMessages Per Person")

for name, count in sorted_people:
    percentage = count / total_messages * 100
    print(f"{name:<15} {count:>5} ({percentage:.1f}%)")
```

```python
from datetime import datetime

# -----------------------------
# MOST ACTIVE DAY & HOUR
# -----------------------------

# Change this format if your chat export is different
date_format = "%d/%m/%y, %H:%M"

day_count = {}
hour_count = {}

for msg in messages:

    # Convert timestamp string to datetime object
    dt = datetime.strptime(msg["timestamp"], date_format)

    # Date only (e.g., 2024-04-14)
    day = dt.date()

    # Hour only (0-23)
    hour = dt.hour

    # Count messages per day
    if day in day_count:
        day_count[day] += 1
    else:
        day_count[day] = 1

    # Count messages per hour
    if hour in hour_count:
        hour_count[hour] += 1
    else:
        hour_count[hour] = 1

# Find busiest day
busiest_day = max(day_count, key=day_count.get)
busiest_day_messages = day_count[busiest_day]

# Find busiest hour
busiest_hour = max(hour_count, key=hour_count.get)
busiest_hour_messages = hour_count[busiest_hour]

# Number of days in the chat
total_days = len(day_count)

# Average messages during that hour per day
average = busiest_hour_messages / total_days

# -----------------------------
# PRINT REPORT
# -----------------------------

print("=" * 60)
print("MOST ACTIVE DAY & HOUR".center(60))
print("=" * 60)

print(f"Busiest day  : {busiest_day.strftime('%d %B %Y')} ({busiest_day_messages} messages)")
print(f"Busiest hour : {busiest_hour:02d}:00 - {(busiest_hour + 1) % 24:02d}:00 ({average:.1f} avg messages/day)")
```

```python
import string

print("=" * 60)
print("TOP 10 MOST USED WORDS")
print("=" * 60)

# Common stop words
stop_words = [
    "i", "is", "the", "a", "an", "and", "or", "to",
    "of", "in", "on", "for", "at", "it", "this",
    "that", "you", "me", "my", "we", "our", "your",
    "are", "was", "were", "be", "am", "with",
    "as", "by", "from", "if", "but", "so",
    "have", "has", "had", "do", "does", "did"
]

word_count = {}

for msg in messages:

    text = msg["message"]

    # Skip media messages
    if "<Media omitted>" in text:
        continue

    # Skip deleted messages
    if "deleted this message" in text.lower():
        continue

    # Split message into words
    words = text.split()

    for word in words:

        # Lowercase
        word = word.lower()

        # Remove punctuation
        word = word.strip(string.punctuation)

        # Ignore empty words
        if word == "":
            continue

        # Ignore stop words
        if word in stop_words:
            continue

        # Count words
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

# Sort dictionary by frequency
top_words = sorted(
    word_count.items(),
    key=lambda x: x[1],
    reverse=True
)

print()

maximum = top_words[0][1]

for word, count in top_words[:10]:

    # Bar length (maximum 30 blocks)
    bar_length = int((count / maximum) * 30)

    bar = "█" * bar_length

    print(f"{word:<15} {count:>4}  {bar}")
```

```python
from datetime import datetime, timedelta

print("=" * 70)
print("FEATURE 6 - RESPONSE SPEED & SILENT STREAKS")
print("=" * 70)

# Convert timestamp strings into datetime objects
for msg in messages:
    msg["datetime"] = datetime.strptime(msg["timestamp"], "%d/%m/%y, %H:%M")

# Sort messages by time
messages.sort(key=lambda x: x["datetime"])

participants = sorted(list(participants))

# =====================================================
# PART A - AVERAGE RESPONSE TIME
# =====================================================

response_times = {}

for person in participants:
    response_times[person] = []

for i in range(1, len(messages)):

    previous = messages[i - 1]
    current = messages[i]

    # Ignore consecutive messages by the same person
    if previous["sender"] == current["sender"]:
        continue

    gap = current["datetime"] - previous["datetime"]

    minutes = gap.total_seconds() / 60

    response_times[current["sender"]].append(minutes)

average_response = {}

print("\nRESPONSE PATTERNS\n")

for person in participants:

    gaps = response_times[person]

    if len(gaps) == 0:
        average = 0
    else:
        average = sum(gaps) / len(gaps)

    average_response[person] = average

    if average == 0:
        print(f"{person:<10} : No replies")

    elif average < 60:
        print(f"{person:<10} : {average:.2f} minutes")

    else:
        print(f"{person:<10} : {average/60:.2f} hours")

# Find fastest and slowest replier
valid = {k: v for k, v in average_response.items() if v > 0}

fastest = min(valid, key=valid.get)
slowest = max(valid, key=valid.get)

print("\nFastest Replier :", fastest)

if valid[fastest] < 60:
    print(f"Average Response : {valid[fastest]:.2f} minutes")
else:
    print(f"Average Response : {valid[fastest]/60:.2f} hours")

print()

print("Slowest Replier :", slowest)

if valid[slowest] < 60:
    print(f"Average Response : {valid[slowest]:.2f} minutes")
else:
    print(f"Average Response : {valid[slowest]/60:.2f} hours")

# =====================================================
# PART B - LONGEST SILENT STREAK
# =====================================================

print("\n")
print("=" * 70)
print("LONGEST SILENT STREAKS")
print("=" * 70)

person_dates = {}

for person in participants:
    person_dates[person] = set()

all_dates = []

for msg in messages:

    date = msg["datetime"].date()

    all_dates.append(date)

    person_dates[msg["sender"]].add(date)

start_date = min(all_dates)
end_date = max(all_dates)

for person in participants:

    longest = 0
    current = 0

    longest_start = None
    longest_end = None
    streak_start = None

    day = start_date

    while day <= end_date:

        if day not in person_dates[person]:

            if current == 0:
                streak_start = day

            current += 1

        else:

            if current > longest:
                longest = current
                longest_start = streak_start
                longest_end = day - timedelta(days=1)

            current = 0

        day += timedelta(days=1)

    # Check if streak continues until last day
    if current > longest:
        longest = current
        longest_start = streak_start
        longest_end = end_date

    if longest == 0:
        print(f"{person:<10} : 0 days (never went silent)")
    else:
        print(
            f"{person:<10} : {longest} days "
            f"({longest_start.strftime('%d %b')} to {longest_end.strftime('%d %b')})"
        )
```
